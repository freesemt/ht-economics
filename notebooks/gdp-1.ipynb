```python
import sys
sys.path.insert(0, '..')
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['font.family'] = ["MS Gothic"]  # "MS Gothic", "MS Mincho", "Meiryo"
from ht_economics.GDP import get_gdp_data
```

```python
url_j = "https://www.esri.cao.go.jp/jp/sna/data/data_list/sokuhou/files/2024/qe244_2/tables/ritu-jcy2442.csv" # 年次GDP成長率（実質）
url_m = "https://www.esri.cao.go.jp/jp/sna/data/data_list/sokuhou/files/2024/qe244_2/tables/ritu-mcy2442.csv" # 年次GDP成長率（名目）
gdp_data_j = get_gdp_data(url_j)
gdp_data_j.head(3)
```

```python
gdp_data_m = get_gdp_data(url_m)
gdp_data_m.head(3)
```

```python
plt.plot(gdp_data_j["Year"], gdp_data_j["GDP"], label="実質")
plt.plot(gdp_data_m["Year"], gdp_data_m["GDP"], label="名目");
plt.xlabel("年")
plt.ylabel("成長率 (%)")
plt.legend();
```

```python
from datetime import timedelta
width = timedelta(days=200)
shift = timedelta(days=100)
plt.bar(gdp_data_j["Year"], gdp_data_j["GDP"], alpha=0.5, width=width,label="実質")
plt.bar(gdp_data_m["Year"]+shift, gdp_data_m["GDP"], alpha=0.5, width=width,label="名目");
plt.xlabel("年")
plt.ylabel("成長率 (%)")
plt.legend();
```
